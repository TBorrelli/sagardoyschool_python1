<h1 id="tuple">Tuples</h2>


Las tuplas de Python son un tipo de estructura de datos muy similar a las listas.<br>
La principal diferencia entre ambas es que las tuplas son inmutables, lo que significa que no pueden modificarse una vez creadas. 
Esto las hace ideales para almacenar datos que no deben ser modificados, <b>como los registros de una base de datos.</b>

En Python, existen diferentes tipos de datos: <code>string</code>, <code>integer</code>, <code>float</code> y <code>boolean</code>. <br>
Todos estos tipos de datos pueden estar contenidos en una tupla

Cada elemento de la tupla, incluidas otras tuplas, puede obtenerse mediante un índice, como se muestra en la figura:

````python
# Imprimir elemento en cada colomna
print("Elemento 0 de tupla: ", tupla_anidada[0])
print("Elemento 1 de tupla: ", tupla_anidada[1])
print("Elemento 2 de tupla: ", tupla_anidada[2])
print("Elemento 3 de tupla: ", tupla_anidada[3])
````

Podemos utilizar el segundo índice (linea) para acceder a otras tuplas:

````python
#Columna 0
print("Elemento 0, 0 de tupla: ", tupla_anidada[0][0])
print("Elemento 0, 1 de tupla: ", tupla_anidada[0][1])
print("Elemento 0, 2 de tupla: ", tupla_anidada[0][2])

#Columna 1
print("Elemento 1, 0 de tupla: ", tupla_anidada[1])

#Columna 2
print("Elemento 2, 0 de tupla: ", tupla_anidada[2])

#Columna 3
print("Elemento 3, 0 de tupla: ", tupla_anidada[3][0])
print("Elemento 3, 1 de tupla: ", tupla_anidada[3][1])
print("Elemento 3, 2 de tupla: ", tupla_anidada[3][2])
````

Podemos utilizar el terzero índice (linea - de linea) [columna 3, linea 1] para acceder a otras tuplas:

````python
#Columna 0
print("Elemento 0, 0 de tupla: ", tupla_anidada[0][0])
print("Elemento 0, 1 de tupla: ", tupla_anidada[0][1])
print("Elemento 0, 2 de tupla: ", tupla_anidada[0][2])

#Columna 1
print("Elemento 1, 0 de tupla: ", tupla_anidada[1])

#Columna 2
print("Elemento 2, 0 de tupla: ", tupla_anidada[2])

#Columna 3
print("Elemento 3, 0 de tupla: ", tupla_anidada[3][0])
print("Elemento 3, 1, 0 de tupla: ", tupla_anidada[3][1][0])
print("Elemento 3, 1, 1 de tupla: ", tupla_anidada[3][1][1])
print("Elemento 3, 1, 2 de tupla: ", tupla_anidada[3][1][2])
print("Elemento 3, 2 de tupla: ", tupla_anidada[3][2])
````

Podemos utilizar el cuarto índice (de cada string) para acceder a cada caractere de las string:

````python
#...
print("Elemento 3, 1, 0 de tupla: ", tupla_anidada[3][1][0][:3])
print("Elemento 3, 1, 0 de tupla: ", tupla_anidada[3][1][0][4:7])
print("Elemento 3, 1, 0 de tupla: ", tupla_anidada[3][1][0][8:])
#...
````
