<h1>Operadores lógicos</h1>

A veces desea verificar más de una condición a la vez. Por ejemplo, es posible que desee comprobar si una condición y otra condición son **True**. <br>
Los operadores lógicos le permiten combinar o modificar condiciones.
<ul>
 <li><code>and</code></li>
 <li><code>or</code></li>
 <li><code>not</code></li>
</ul>

Estos operadores se resumen para dos variables utilizando las siguientes tablas de verdad:

<h2><code>and</code></h2>

<img src="data:image/png;base64, " align="center" />

La declaración <code>and</code> sólo es **True** cuando ambas condiciones son verdaderas.

<h2><code>or</code></h2>

<img src="data:image/png;base64, " align="center" />

 La declaración <code>or</code> es **True** si una condición, o ambas son **True**.

<h2><code>not</code></h2>

La declaración <code>not</code> produce el valor de verdad opuesto <br>
<ul>
    <li><code>True -> False</code></li>
    <li><code>False -> True</code></li>
</ul>

<h2>Ejemplos</h2>

Veamos cómo determinar si un álbum se publicó después de 1979 (no se incluye 1979) <code>and</code> antes de 1990 (no se incluye 1990). <br>
Los periodos comprendidos entre 1980 <code>and</code> 1989 cumplen esta condición. 

````python
álbum_año = 1980

if (álbum_año > 1979) and (álbum_año < 1990):
    print ("El año del álbum estuvo comprendido entre 1980 y 1989")
````

Para determinar si un álbum se publicó antes de 1980 (1979 y años anteriores) o después de 1989 (1990 y años posteriores), puede utilizarse la expresión <code>or</code><br>
Los periodos anteriores a 1980 (1979 y anteriores) <code>or</code> posteriores a 1989 (1990 y posteriores) cumplen esta condición.

````python
álbum_año = 1990

if(álbum_año < 1980) or (álbum_año > 1989):
    print ("El álbum no se fabricó en la década de 1980")
else:
    print("El Álbum se hizo en los años 80")
````

La declaración <code>not</code> comprueba si la sentencia es falsa:

````python
álbum_año = 1983

if not (álbum_año == 1984):
    print ("El año del álbum no es 1984")
````
