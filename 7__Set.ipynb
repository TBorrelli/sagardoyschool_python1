<h1 id="set">Sets</h1>


Un <code>set</code> (conjunto) es una colección única de objetos en Python. <br>
Puedes denotar un conjunto con un par de llaves <b>{}</b>. <br>
Python eliminará automáticamente los elementos duplicados:

````python
set_ciudades = {"Roma", "Madrid", "Valencia", "Milano", "Roma", "Alicante", "Valencia", "Madrid"}
print(set_ciudades)
````

También puede crear un <code>set</code> a partir de una lista del siguiente modo:

````python
lista_ciudades = ["Roma", "Madrid", "Valencia", "Milano", "Roma", "Alicante", "Valencia", "Madrid"]
set_ciudades = set(lista_ciudades)             
print(set_ciudades)
````

<h2 id="op">Operaciones de set</h2>

Repasemos las operaciones con <code>set</code>, ya que pueden utilizarse para modificar el <code>set</code>. Consideremos el <code>set</code> <b>set_a</b>:

````python
set_a = {"Roma", "Madrid", "Valencia", "Milano", "Roma", "Alicante", "Valencia", "Madrid"}
# o
set_a = set(["Roma", "Madrid", "Valencia", "Milano", "Roma", "Alicante", "Valencia", "Madrid"])
print(set_a)
````

Podemos añadir un elemento a un conjunto utilizando el método <code>add()</code>:

````python
set_a.add("Toledo")
set_a
````

Si añadimos el mismo elemento dos veces, no pasará nada, ya que no puede haber duplicados en un <code>set</code>:

````python
set_a.add("Toledo")
set_a
````

Podemos eliminar un elemento de un <code>set</code> utilizando el <code>remove</code> método:

````python
set_a.remove("Toledo")
set_a
````

Podemos verificar si un elemento está en el <code>set</code> utilizando el <code>in</code> método:

````python
"Valencia" in set_a
````

````python
"Barcelona" in set_a
````

<h2 id="lógica">Operaciones lógicas con set</h2>

Recuerda que con los <code>set</code> puedes comprobar la diferencia entre <code>set</code>, así como la diferencia simétrica, la intersección y la unión:

Consideremos los dos <code>set</code> siguientes:

````python
# Sample Sets

set_ciudades_espana = set(["Madrid", 'Alicante', 'Leon',"Cartagena"])
set_ciudades_nicaragua = set([ "Managua", "Leon", "Chinandega", "Cartagena"])
````

<img src="data:image/png;base64, " width="600" align="center" />

````python
set_ciudades_espana, set_ciudades_nicaragua
````

Como ambos <code>set</code>contienen <b>Leon</b> y <b>Cartagena</b> representamos estos elementos comunes con la intersección de dos círculos

<img src="data:image/png;base64, " width="600" align="center" />

<h3>Intersección</h3>

Puedes encontrar la intersección de dos <code>set</code> como sigue usando <code>&</code> o a través el metodo <code>intersection</code>:

````python
intersección = set_ciudades_espana & set_ciudades_nicaragua
# o
set_ciudades_espana.intersection(set_ciudades_nicaragua)   
````

<img src="data:image/png;base64, " width="600" align="center" />

<h3>Diferencia</h3>

Puedes encontrar todos los elementos que solo están contenidos en <code>set_ciudades_espana</code> utilizando el método <code>difference</code>:

````python
set_ciudades_espana.difference(set_ciudades_nicaragua)
````

Sólo hay que tener en cuenta los elementos de <code>set_ciudades_espana</code>; todos los elementos en <code>set_ciudades_nicaragua</code>, incluida la intersección, no están incluidos.

<img src="data:image/png;base64, " width="600" align="center" />

Los elementos en <code>set_ciudades_nicaragua</code> pero no en <code>set_ciudades_espana</code> viene dado por:

````python
set_ciudades_nicaragua.difference(set_ciudades_espana)  
````

<img src="data:image/png;base64, " width="600" align="center" />

<h3>Unión</h3>

La unión corresponde a todos los elementos de ambos <code>set</code>, lo que se representa coloreando ambos círculos. <br>
La unión viene dada con el método <code>union</code>:

````python
set_ciudades_espana.union(set_ciudades_nicaragua)
````

<img src="data:image/png;base64, " width="600" align="center" />

<h3>Superconjunto o subconjunto</h3>

Y puedes comprobar si un conjunto es un <code>superset</code> (superconjunto) o <code>subset</code> (subconjunto) de otro <code>set</code>, respectivamente, de la siguiente manera:

````python
algunas_ciudades_nicaragua = {"Leon","Managua"}
````

````python
# superconjunto
set_ciudades_nicaragua.issuperset(algunas_ciudades_nicaragua)   
````

````python
# subconjunto
algunas_ciudades_nicaragua.issubset(set_ciudades_nicaragua)  
````
